```python
data = pd.read_json('../data/mrbilit_search.json')
test_data = pd.read_json('../data/test_data.json')
iran_cities = pd.read_csv('../data/iran_cities.csv')
typo_char = pd.read_csv('../data/typo_char.csv')
top_cities = pd.read_json('../data/top_cities.json')
top_provinces = pd.read_json('../data/top_provinces.json')
```

```python
# Convert English words to Farsi

EN_FA_type = typo_char.set_index('EN')['FA'].to_dict()

CityEN_CityFA = iran_cities.iloc[:,0:2]
CityEN_CityFA = CityEN_CityFA.set_index('City EN')['City FA'].to_dict()

def en_to_fa(words):
  farsi = ""
  if words in CityEN_CityFA:
    farsi = CityEN_CityFA[words]
    return farsi

  for word in  words:
    farsi = farsi + EN_FA_type.get(word.lower(), word)
  return farsi

print(en_to_fa('Sghl من nhKdhg هستم'))
print(en_to_fa('Zibashahr'))
```

```python
data = data[(data['ServiceType'] == 'taxi') | (data['ServiceType'] == 'bus')]
data.reset_index(drop = True, inplace =True)
data = data.iloc[:,1:3]
data
```

```python
set_Cities= set(iran_cities['City FA'])
list_top_cities = top_cities[0].tolist()

def finder(words):
  filtered_set = [item for item in set_Cities if str(item).startswith(words)]
  if len(filtered_set) > 5:
    return  [item for item in filtered_set if item in list_top_cities]

  return filtered_set


finder('تهران')
```

```python

# Define a function to filter out empty strings from a list
def filter_empty_strings(lst):
    return [item for item in lst if item != '']

# Apply the function to the 'TypedStrings' column
data['TypedStrings'] = data['TypedStrings'].apply(filter_empty_strings)


# Filter and remove rows with lists containing only empty strings in 'Column1'
data = data[data['TypedStrings'].apply(lambda x: any(item != '' for item in x))]

data.reset_index(drop=True,inplace=True)
data
```

```python
data['AcceptString2'] = data['AcceptString'].apply(lambda x: [x])
data = data[data['TypedStrings'] != data['AcceptString2']]
data.drop(columns='AcceptString2',inplace=True)
data.reset_index(drop=True,inplace=True)
data
```

```python
from collections import Counter

def include(word):
  result = data[data['TypedStrings'].apply(lambda x: word in x )]
  result = result['AcceptString'].tolist()

  element_counts = Counter(result)
  # Sort the elements based on their counts in descending order
  result = sorted(element_counts, key=lambda x: element_counts[x], reverse=True)
  return result

include('تهد')
```

```python
def recommender(words):
  words = en_to_fa(words)
  suggest = include(words)
  fi = finder(words)
  if len(suggest) >= 5:
    return suggest[:5]

  new = suggest + fi + list_top_cities

  unique_items = []
  for item in new:
      if item not in unique_items:
        unique_items.append(item)


  return unique_items[:5]

recommender('ت')
```
